```python
import numpy as np
import scipy as sp
from scipy.stats import beta
import matplotlib.pyplot as plt
```

```python
s = np.random.uniform(0,1,100)
u = np.linspace(0, 1, 100)
fig, ax = plt.subplots(1,2)

ax[0].scatter(range(0,100),s)
ax[0].set_xlim(0,100)
#a
Beta = beta(100,200)
ax[1].plot(u, Beta.pdf(u))
plt.show()
```

```python
fig, ax = plt.subplots(1,2)

mean = 100/300
ax[0].plot(u,Beta.pdf(u))
ax[0].plot(mean*np.ones(100),Beta.pdf(u),color='g')

ax[1].scatter(range(0,100),s)
ax[1].plot(range(0,100),mean*np.ones(100),color='g')

plt.show()
```
