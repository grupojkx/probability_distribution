```python
import numpy as np
import scipy as sp
import matplotlib.pylab as plt
from scipy.special import comb
from scipy.stats import beta
```

```python
# Binomial distribution
def Bi(m,N,p):
    com = comb(N, m, exact=False)
    return com*p**m*(1-p)**(N-m)
def V_Bi(N,p):
    DB =np.zeros((1, N+1))
    for m in range(0, N+1):
        DB[0,m] = Bi(m,N,p)
    return DB
```

```python
plt.figure()
for i, p in enumerate([0.1,0.15,0.2,0.25,0.3,0.35,0.4,0.45,0.5]):
    plt.subplot(3,3,i+1)
    plt.xlim(0,10)
    plt.ylim(0,0.4)
    plt.scatter(range(0,11),V_Bi(10,p))
    plt.annotate("p={}".format(p), (5, 0.25))
    
plt.show()
```

```python
# Beta distribution
u = np.linspace(0, 1, 100)
for i, [a, b] in enumerate([[0.002, 0.002], [3, 2], [4, 2], [15, 2]]):
    plt.subplot(2, 2, i + 1)
    Beta = beta(a, b)
    plt.xlim(0, 1)
    plt.ylim(0, 6.4)
    plt.plot(u, Beta.pdf(u))
    plt.annotate("a={}".format(a), (0.1, 2.7))
    plt.annotate("b={}".format(b), (0.1, 2.0))
plt.show()
```
